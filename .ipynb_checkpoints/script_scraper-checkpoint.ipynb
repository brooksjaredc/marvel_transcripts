```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://transcripts.fandom.com/wiki/Category:Marvel_Transcripts'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tags = soup.findAll('a', {"class": "category-page__member-link"})
links = ['https://transcripts.fandom.com' + a['href'] for a in tags]
links[4]
```

```python
response1 = requests.get(links[4])
```

```python
soup = BeautifulSoup(response1.text, 'html.parser')
# script = soup.find('div', {'class': 'mw-content-ltr mw-content-text'})
soup.find('h1', {'class': 'page-header__title'}).get_text()
```

```python
links[12]
response1 = requests.get(links[12])
soup = BeautifulSoup(response1.text, 'html.parser')
soup.find('div', {'class': 'mw-content-ltr mw-content-text'}).findAll('p')
```

```python
links[4]
response1 = requests.get(links[12])
soup = BeautifulSoup(response1.text, 'html.parser')
soup.find('div', {'class': 'mw-content-ltr mw-content-text'}).get_text().strip()
```

```python
scripts = []
titles = []

for link in links:
    response1 = requests.get(link)
    soup = BeautifulSoup(response1.text, 'html.parser')
    script = soup.find('div', {'class': 'mw-content-ltr mw-content-text'}).get_text().strip()
#     if script.find('ul') is None:
#         continue
#     script = script.find('ul').get_text()
#     script = script.findAll('p').get_text()
    title = soup.find('h1', {'class': 'page-header__title'}).get_text()
    if len(script) > 1000:
        scripts.append(script)
        titles.append(title)
```

```python
scripts[1][:100]
titles[8]
len(titles)
```

```python
import pandas
df = pandas.DataFrame(data={"titles": titles, "scripts": scripts})
df.to_csv("./scripts_and_titles.csv", sep=',',index=False)
```
